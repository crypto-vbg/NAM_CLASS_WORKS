```python
from nltk.corpus import gutenberg
print(gutenberg.fileids())
```

```python
nb_data = gutenberg.raw("shakespeare-macbeth.txt")
print(nb_data)
```

```python
characters = ['Macbeth', 'Lady_Macbeth', 'The_Three_Witches', 'Banquo', 'King_Duncan', \
'Macduff', 'Malcolm', 'Hecate', 'Fleance', 'Lennox', 'Ross', 'The_Murderers', \
'Porter', 'Lady_Macduff', 'Donalbain']
```

```python
nb_data = nb_data.replace('Macb.','Macbeth')
nb_data = nb_data.replace('The Three Witches','The_Three_Witches')
nb_data = nb_data.replace('King.','King_Duncan')
nb_data = nb_data.replace('The Murderers','The_Murderers')
nb_data = nb_data.replace('Lady.','Lady_Macduff')
nb_data = nb_data.replace('Banq.','Banquo')
nb_data = nb_data.replace('Rosse.','Ross')
```

```python
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(nb_data)
print(sentences)
print(len(sentences))
```

```python
for sent in sentences:
    if("Scena" in sent or "Actus" in sent or "Scoena" in sent):
        sentences.remove(sent)
    elif ("William Shakespeare" in sent):
        sentences.remove(sent)

print(len(sentences))
```

```python
import pandas as pd
adj_data = pd.DataFrame(0,index=characters , columns = characters)
adj_data.head(len(characters))
```

```python
for sentence in sentences:
    for char1 in characters:
        for char2 in characters:
            if(char1 != char2)and char1 in sentence and char2 in sentence:
                adj_data[char1][char2]+=1
                adj_data[char2][char1]+=1
adj_data.head(len(characters))
```

```python
import matplotlib.pyplot as plt
plt.figure(2,(24,24))
G = nx.from_pandas_adjacency(adj_data)
pos = nx.spring_layout(G,seed = 42)
nx.draw_networkx(G,pos)
nx.draw_networkx_edges(G,pos)

edge_labels = nx.get_edge_attributes(G,"weight")
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels)
plt.show()
```
