```python
import networkx as nx
```

```python
print(nx.__version__)
```

```python
G = nx.Graph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")

pos = nx.spring_layout(G)
nx.draw(G,pos,node_color = "green" , node_size = 800 , with_labels = True)
```

```python
#adding edges
G.add_edge("A" , "B")
G.add_edge("B" , "C")
G.add_edge("C" , "D")
G.add_edge("D" , "A")

nx.draw(G,pos,node_color = "green" , node_size = 800 , with_labels = True)
```

```python
#degree
G.degree
```

# Directed Graph

```python
G = nx.DiGraph()
G.add_edge("A" , "B")
G.add_edge("C" , "B")
G.add_edge("D" , "B")
G.add_edge("A" , "D")
nx.draw(G,pos,node_color = "green" , node_size = 800 , with_labels = True)
```

```python
#in-degree
G.in_degree()
```

```python
nx.has_path(G,"C","A")
```

```python
nx.shortest_path(G,"A","B")
```

```python
list(nx.all_simple_paths(G,"A","B"))
```

# Weighted Graph

```python
G = nx.Graph()
G.add_edge("A" , "B" , weight = 0.6)
G.add_edge("C" , "B", weight = 0.8)
G.add_edge("D" , "B", weight = 0.3)
G.add_edge("A" , "D", weight = 0.2)
nx.draw(G,pos,node_color = "green" , node_size = 800 , with_labels = True)
```

```python
elarge = [(u,v) for (u,v,d) in G.edges(data=True) if d["weight"]>0.5]
esmall = [(u,v) for (u,v,d) in G.edges(data=True) if d["weight"]<=0.5]
print(elarge)
print(esmall)
```

```python
pos = nx.spring_layout(G, seed=42)
#render nodes
nx.draw_networkx_nodes(G,pos)
#render node labels
nx.draw_networkx_labels(G,pos)
#render edges
nx.draw_networkx_edges(G,pos,edgelist =elarge , width = 6)
nx.draw_networkx_edges(G,pos,edgelist =esmall , width = 6,alpha = 0.5 , style= "dashed")
#render edge labels
edge_lists = nx.get_edge_attributes(G,"weight")
nx.draw_networkx_edge_labels(G,pos,edge_labels = edge_lists)
```
