```python
import networkx as nx
```

```python
g = nx.disjoint_union(nx.petersen_graph() , nx.tetrahedral_graph());
nx.draw(g , with_labels = True)
```

```python
list(nx.connected_components(g))
```

```python
dg = nx.disjoint_union(nx.tetrahedral_graph(create_using=nx.DiGraph()),nx.tetrahedral_graph(create_using=nx.DiGraph()))
dg.add_edge(4,0)
nx.draw(dg,with_labels = True)
```

```python
list(nx.strongly_connected_components(dg))
```

```python
list(nx.weakly_connected_components(dg))
```

# Cliques

```python
kg = nx.karate_club_graph()
nx.draw(kg , with_labels = True)
```

```python
cliques = list(nx.find_cliques(kg))
[clique for clique in cliques if(len(clique)>2)]
```

```python
sb = kg.subgraph([0,1,2,3,7])
nx.density(sb)
```

# K-core

```python
nx.core_number(kg)
```

```python
k4 = nx.k_core(kg,4)
nx.draw(k4,with_labels = True)
```

```python
from networkx.algorithms.community import k_clique_communities
```

```python
list(k_clique_communities(kg,4))
```
